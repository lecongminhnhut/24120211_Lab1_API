```python
!sudo apt update
!sudo apt install -y pciutils
!apt-get update -qq
!apt-get install -y zstd
!curl -fsSL https://ollama.com/install.sh | sh

import os
env = os.environ.copy()
env["OLLAMA_HOST"] = "0.0.0.0"
env["OLLAMA_ORIGINS"] = "*"

import subprocess
def run_ollama_serve():
  subprocess.Popen(["ollama", "serve"], env=env)

import threading
thread = threading.Thread(target=run_ollama_serve)
thread.start()

import time
time.sleep(5)
```

```python
# Định nghĩa Modelfile cho mô hình mới Ollama-Pinggy.ipynb]
modelfile = """
FROM gemma2:2b
PARAMETER temperature 0.1
SYSTEM \"\"\"
Bạn là một trợ lý AI chuyên nghiệp của sinh viên.
Nhiệm vụ: Phân loại văn bản/email được cung cấp.
- Nếu nội dung là quảng cáo, lừa đảo, trúng thưởng, mời chào: Trả về 'SPAM'.
- Nếu nội dung là trao đổi học tập, công việc, bạn bè: Trả về 'HAM'.
Chỉ trả về đúng 1 từ duy nhất, không giải thích.
\"\"\"
"""

with open("Modelfile", "w") as f:
    f.write(modelfile)

# Kéo mô hình gốc và khởi tạo mô hình tùy chỉnh 'artyann-gemma' Ollama-Pinggy.ipynb]
!ollama pull gemma2:2b
!ollama create artyann-gemma -f Modelfile
print("--- Đã tạo xong mô hình artyann-gemma thành công! ---")
```
